#Q1

Write the products management project as follows.
1. Product information includes name and price.
2. Use list to save information
3. We do not have duplicate product name
4. This project includes sections for adding, deleting, editing, detail and searching by product name.
6. For the detail section, display the total number of products, the average price, the most expensive and cheapest product.
5. The program should continue until the user enters "exit".

```python
products = []

def add_product(name, price):
    for product in products:
        if product['name'] == name:
            print('Product already exists')
            return
    products.append({'name': name, 'price': price})
    print('Product added successfully')

def delete_product(name):
    for product in products:
        if product['name'] == name:
            products.remove(product)
            print('Product deleted successfully')
            return
    print('Product not found')

def edit_product(name, price):
    for product in products:
        if product['name'] == name:
            product['price'] = price
            print('Product edited successfully')
            return
    print('Product not found')

def detail():
    total_products = len(products)
    if total_products == 0:
        print('No products found')
        return
    total_price = sum([product['price'] for product in products])
    average_price = total_price / total_products
    most_expensive_product = max(products, key=lambda x: x['price'])
    cheapest_product = min(products, key=lambda x: x['price'])
    
    print(f'Total number of products: {total_products}')
    print(f'Average price: {average_price}')
    print(f'Most expensive product: {most_expensive_product["name"]} ({most_expensive_product["price"]})')
    print(f'Cheapest product: {cheapest_product["name"]} ({cheapest_product["price"]})')

def search(name):
    for product in products:
        if name.lower() in product['name'].lower():
            print(f'{product["name"]}: {product["price"]}')
            
def main():
    while True:
        command = input('Enter command (add/delete/edit/detail/search/exit): ')
        
        if command == 'add':
            name = input('Enter name: ')
            price = float(input('Enter price: '))
            add_product(name, price)
            
        elif command == 'delete':
            name = input('Enter name: ')
            delete_product(name)
            
        elif command == 'edit':
            name = input('Enter name: ')
            price = float(input('Enter new price: '))
            edit_product(name, price)
            
        elif command == 'detail':
            detail()
            
        elif command == 'search':
            name = input('Enter name to search for: ')
            search(name)
            
        elif command == 'exit':
            break
            
    print('Goodbye!')

if __name__ == '__main__':
    main()
```

```python
#user names and password, admin:123, user:145
# this code creates a random 13 digits for a barcode also stores the username of the one who last edited a product
import random


products = []
removed_products = []
users = [{'username': 'admin', 'password': '123', 'role': 'admin'}, {'username': 'user', 'password': '145', 'role': 'user'}, {'username': 'none', 'password': '', 'role': 'none'}]
current_user = None

def login():
    global current_user
    username = input('Enter username: ')
    password = input('Enter password: ')
    for user in users:
        if user['username'] == username and user['password'] == password:
            current_user = user
            print('Login successful')
            return
    print('Invalid username or password')

def logout():
    global current_user
    current_user = None
    print('Logout successful')

def add_product(name: str, price: float) -> dict:
    if current_user['role'] != 'admin':
        print('You do not have permission to add a product')
        return {}
    
    for product in products:
        if product['name'] == name:
            print('Product already exists')
            return {}
    
    new_product = {'name': name, 'price': price, 'barcode': random.randint(1000000000000, 9999999999999), 'last_edited_by': current_user['username']}
    
    products.append(new_product)
    
    print('Product added successfully')
    
    return new_product

def delete_product(name):
    if current_user['role'] != 'admin':
        print('You do not have permission to delete a product')
        return
    
    for product in products:
        if product['name'] == name:
            products.remove(product)
            removed_products.append(product)
            
            print('Product deleted successfully')
            
            return
    
    print('Product not found')

def edit_product(name, price):
    if current_user['role'] != 'admin':
        print('You do not have permission to edit a product')
        return
    
    for product in products:
        if product['name'] == name:
            product['price'] = price
            product['last_edited_by'] = current_user['username']
            
            print('Product edited successfully')
            
            return
    
    print('Product not found')

def detail():
    total_products = len(products)
    
    if total_products == 0:
        print('No products found')
        
        return
    
    total_price = sum([product['price'] for product in products])
    
    average_price = total_price / total_products
    
    most_expensive_product = max(products, key=lambda x: x['price'])
    
    cheapest_product = min(products, key=lambda x: x['price'])
    
    print(f'Total number of products: {total_products}')
    
    print(f'Average price: {average_price}')
    
    print(f'Most expensive product: {most_expensive_product["name"]} ({most_expensive_product["price"]})')
    
    print(f'Cheapest product: {cheapest_product["name"]} ({cheapest_product["price"]})')

def search(name):
    for product in products:
        if name.lower() in product['name'].lower():
            print(f'{product["name"]}: {product["price"]}')

def main():
    while True:
        command = input('Enter command (login/logout/add/delete/edit/detail/search/exit): ')
        
        if command == 'login':
            login()
            
        elif command == 'logout':
            logout()
            
        elif command == 'add':
            if current_user is None or current_user['role'] == 'user':
                print('You do not have permission to add a product')
                
                continue
            
            name = input('Enter name: ')
            
            price = float(input('Enter price: '))
            
            add_product(name, price)
            
        elif command == 'delete':
            if current_user is None or current_user['role'] == 'user':
                print('You do not have permission to delete a product')
                
                continue
            
            name = input('Enter name: ')
            
            delete_product(name)
            
        elif command == 'edit':
            if current_user is None or current_user['role'] == 'user':
                print('You do not have permission to edit a product')
                continue
            
            name = input('Enter name: ')
            
            price = float(input('Enter new price: '))
            
            edit_product(name, price)
            
        elif command == 'detail':
            detail()
            
        elif command == 'search':
            name = input('Enter name: ')
            
            search(name)
            
        elif command == 'exit':
            break
        
        else:
            print('Invalid command')

if __name__ == '__main__':
    main()
```
